```python
import xml.etree.ElementTree as ET
tree = ET.parse('Homo_sapiens_AR109.20190905_annotation_report.xml')
root = tree.getroot()
print(root.tag, root.attrib)
for child in root:
    print(child.tag, child.attrib)
```

```python
# children
count = 0 
for child in root:
    count += 1
print(count)
```

```python
# grandchildren
count = 0
for child in root:
    print('Child:')
    print(child.tag, child.attrib)
    print('Grandchildren:')
    for grandchild in child:
        count += 1
        if count < 10:
            print(grandchild.tag, grandchild.attrib)
    print(count)
    print('\n\n')
```

```python
# great grandchildren
count = 0
for child in root:
    print('Child:')
    print(child.tag, child.attrib)
    print('Grandchildren:')
    for grandchild in child:
        count += 1
        if count < 10:
            print(grandchild.tag, grandchild.attrib)
        for greatgrandchild in grandchild:
            count += 1
            if count < 10:
                print('Great Grandchildren: ', greatgrandchild.tag, greatgrandchild.attrib)
            for greatgreatgrandchild in greatgrandchild:
                count += 1
                if count < 10:
                    print('Great Great Grandchildren: ', greatgreatgrandchild.tag, greatgreatgrandchild.attrib)
                for ggggchild in greatgreatgrandchild:
                    count += 1
                    if count < 10:
                        print('Great Great Great Grandchildren: ', ggggchild.tag, ggggchild.attrib, '\n')
    print(count)
    print('\n\n')
```
